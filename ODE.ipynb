```python
# y' = 1+y^2

h=0.01
min_x = 0.0
max_x = 1.0
n_points = int((max_x - min_x)/h)


x = arange(min_x, max_x, h)
y = zeros(n_points)

x[0] = min_x
y[0] = 0

def func_prime(x,y):
    return (1+y**2)

for i in range(1, n_points):
    deriv = func_prime(x[i-1], y[i-1])    
    y[i] = y[i-1] + h*deriv
    

scatter(x,y)
plot(x,tan(x))



```
